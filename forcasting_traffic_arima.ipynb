```python
dates=list(traffic_data.columns[1:])
views=list(traffic_data.iloc[100,1:])
d={'dates':dates,'views':views}
data=pd.DataFrame(d)
data.head()
```

```python
plt.plot(data['views'])
plt.xticks(rotation=40)
plt.xlabel('day')
plt.ylabel('views')
plt.show()
```

### Spliting the data

```python
train_data=data.iloc[:400,:]
test_data=data.iloc[400:,:]
```

### Checking if it is stationary or not

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adfuller_test(views):
    result=adfuller(views)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    return result[1]
r=adfuller_test(train_data['views'])
```

```python
if r <= 0.05:
    print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data is stationary")
else:
    print("weak evidence against null hypothesis,indicating it is non-stationary ")
```

### Correlation

### ARIMA

```python
from statsmodels.tsa.arima_model import ARIMA
model=ARIMA(train_data['views'],order=(16,0,8))
model_fit=model.fit()
model_fit.summary()
```

```python
test_values=model_fit.predict(start=400,end=549,dynamic=True)
```

```python
plt.plot(test_values)
plt.plot(test_data['views'],color='red')
```
